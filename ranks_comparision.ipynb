# Rank comparison

```python
import json
import matplotlib.pyplot as plt
```

## Comparison ClassicLinkPredNet vs. VectorReconstructionNet

```python
with open('data/ranks_model_ClassicLinkPredNet_derived_bs128_hidden100_bnFalse_d0.json') as json_in:
    ranks_c = json.load(json_in)
    
with open('data/ranks_model_VectorReconstructionNet_derived_bs1000_hidden100_bnFalse_d0.json') as json_in:
    ranks_vr = json.load(json_in)
```

```python
plt.hist(ranks_c, bins=100, range=(0, 1000), label="ClassicLinkPredNet", alpha=0.5)
plt.hist(ranks_vr, bins=100, range=(0, 1000), label="VectorReconstructionNet", alpha=0.5)
plt.ylim([0, 12000])
plt.xlabel("ranks")
plt.ylabel("frequency")
plt.title("Ranks histogram (100 bins, ranks 0-1000)")
plt.legend()

plt.show()
```

## Comparision VectorReconstructionNet trail prediciton vs. head prediction 

The above plot shows that the VectorReconstructionNet has a drop in the frequency of MRR scores at around 10-20. This drop might be caused due to the two different models used for this network (head prediction net and tail prediciton net). Therefore, the following plot shows the ranks of the two models. 

```python
with open('data/ranks_model_VectorReconstructionNet_derived_bs1000_hidden100_bnFalse_d0_head_pred.json') as json_in:
    ranks_head = json.load(json_in)
    
with open('ranks_model_VectorReconstructionNet_derived_bs1000_hidden100_bnFalse_d0_tail_pred.json') as json_in:
    ranks_tail = json.load(json_in)
    
plt.hist(ranks_head, bins=100, range=(0, 500), label="head", alpha=0.5)
plt.hist(ranks_tail, bins=100, range=(0, 500), label="tail", alpha=0.5)
plt.ylim([0, 10000])
plt.xlabel("ranks")
plt.ylabel("frequency")
plt.title("Ranks histogram (100 bins, ranks 0-500)")
plt.legend()

plt.show()
```

```python
The plot shows that both models have the same drop. The reason must be somewhere else. 
```
